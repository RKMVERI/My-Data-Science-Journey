```python
df3 = pd.merge(df1, df2, on = "movie id")
df3
```

```python
# 5
```

```python
genre = df3["genres"].str.get_dummies().columns
genre
```

```python
# 6 # (i)
```

```python
df4 = df3["genres"].str.get_dummies()
df4
```

```python
# 6 # (ii)
```

```python
inter = df3["genres"].str.split("|")
```

```python
collect = {}
for gen in genre:
    l = []
    for i in inter:
        if gen in i:
            l.append(1)
        else:
            l.append(0)
    collect[gen] = l
collect
```

```python
pd.DataFrame(collect)
```

```python
def func1(df, col):
    temp = df["genres"].str.get_dummies()
    return df[temp[col] == 1].groupby("user id").mean()["rating"]
```
