```python
import numpy as np
import pandas as pd
from scipy.optimize import minimize
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
pd.set_option('display.float_format', lambda x: '%.5f' % x)
```

```python
def Ros_fun(x):
     return 100*(x[1] - x[0]**2)**2 + (1 - x[0])**2


def der_Ros_fun(x):
    return np.array([-400*(x[1] - x[0]**2)*(x[0]) - 2*(1 - x[0]), 200*(x[1] - x[0]**2)])


def hes_Ros_fun(x):
    return np.array([[800*(x[0]**2) - 400*(x[1] - x[0]**2) + 2, -400*x[0]], [-400*x[0], 200]])
```

```python
Nfeval = 1
iters = []
a = np.array([0,0])

def callbackF(Xi):
    global Nfeval
    global a
    print('{0:4d}   {1: 3.6f}   {2: 3.6f}   {3: 3.6f}   {4: 3.6f}'.format(Nfeval, Xi[0], Xi[1], Ros_fun(Xi),np.dot(der_Ros_fun(Xi),a)))
    Nfeval += 1
    iters.append(Xi)
    a = der_Ros_fun(Xi)
```

```python
df = minimize(fun=Ros_fun, x0=[-1.2, 1], jac=der_Ros_fun, method='L-BFGS-B', callback=callbackF)
df
```

```python
# For x0=[-1.2, 1]
# 83 iters in Newton CG
# 36 iters in L-BFGS-B
```

```python
Nfeval = 1
iters = []
a = np.array([0,0])

def callbackN(Xi):
    global Nfeval
    global a
    print('{0:4d}   {1: 3.6f}   {2: 3.6f}   {3: 3.6f}   {4: 3.6f}'.format(Nfeval, Xi[0], Xi[1], Ros_fun(Xi),np.dot(der_Ros_fun(Xi),a)))
    Nfeval += 1
    iters.append(Xi)
    a = der_Ros_fun(Xi)
```

```python
minimize(fun=Ros_fun, x0=[-1.2, 1], jac=der_Ros_fun,hess=hes_Ros_fun, method='newton-cg', callback=callbackN)
```
