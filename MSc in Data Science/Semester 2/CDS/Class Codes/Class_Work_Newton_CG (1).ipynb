```python
import numpy as np
import pandas as pd
from scipy.optimize import minimize
import matplotlib.pyplot as plt
from scipy.optimize import fmin_ncg
```

# Quadratic

```python
def fquad(x):
    return (x[0]**4) + (x[0]*x[1]) + ((1 + x[1])**2)
```

```python
def fquad_grad(x):
    d_x1 = 4*(x[0]**3) + x[1]
    d_x2 = x[0] + 2*(1 + x[1])
    grad = np.array([d_x1, d_x2])
    
    return grad
```

```python
def fquad_hess(x):
    d2_x1_x1 = 12*(x[0]**2)
    d2_x1_x2 = 1
    d2_x2_x1 = 1
    d2_x2_x2 = 2
    hess = np.array([[d2_x1_x1, d2_x1_x2], [d2_x2_x1, d2_x2_x2]])
    
    return hess
```

```python
x1 = np.array([5,6])
```

```python
x1
```

```python
f_x = fquad(x1)
d_f_x = fquad_grad(x1)
d2_f_x = fquad_hess(x1)
```

```python
d2_f_x
```

```python
res = minimize(fquad, x1, method='Newton-CG',
               jac = fquad_grad, hess = fquad_hess,
               options = {'xtol': 1e-8, 'disp': True})
```

```python
fig, ax = plt.subplots()

a = np.linspace(-3, 3, 700)
b = np.linspace(-4, 4, 700)

A, B = np.meshgrid(a, b)
c = np.array([A, B])

C = fquad(c)

ax.contourf(A, B, C)
ax.contour(A, B, C, colors = "black")
```

```python
iter_n = 1
a = np.array([0,0])
output = []
def callBack(x):
    global output
    global iter_n
    global a
    output.append([iter_n, x[0], x[1], fquad(x), np.dot(fquad_grad(x),a)])
    iter_n += 1
    a = fquad_grad(x)
```

```python

fmin_ncg(f = fquad, x0 = x1, fprime = fquad_grad, fhess = fquad_hess, maxiter = 80, callback = callBack)

df = pd.DataFrame(output, columns = ['iteration','x1','x2','f(x1,x2)', 'gradient product'])
df

```

```python
x, y = np.meshgrid(np.linspace(-3,3,100),np.linspace(-6,6,100))
z = fquad([x,y])
plt.contour(x, y, z, 70, cmap = 'jet')
plt.plot(df.iloc[:,1],df.iloc[:,2],'r-')
```

# Non Quadratic

```python
def rosen(x):
    """The Rosenbrock function"""
    return sum(100.0*(x[1:]-x[:-1]**2.0)**2.0 + (1-x[:-1])**2.0)
```

```python
def rosen_der(x):
    xm = x[1:-1]
    xm_m1 = x[:-2]
    xm_p1 = x[2:]
    der = np.zeros_like(x)
    der[1:-1] = 200*(xm-xm_m1**2) - 400*(xm_p1 - xm**2)*xm - 2*(1-xm)
    der[0] = -400*x[0]*(x[1]-x[0]**2) - 2*(1-x[0])
    der[-1] = 200*(x[-1]-x[-2]**2)
    return der
```

```python
def rosen_hess(x):
    x = np.asarray(x)
    H = np.diag(-400*x[:-1],1) - np.diag(400*x[:-1],-1)
    diagonal = np.zeros_like(x)
    diagonal[0] = 1200*x[0]**2-400*x[1]+2
    diagonal[-1] = 200
    diagonal[1:-1] = 202 + 1200*x[1:-1]**2 - 400*x[2:]
    H = H + np.diag(diagonal)
    return H
```

```python
x2 = np.array([1,2,-1.08,0.998,0.7,-0.6])
```

```python
res = minimize(rosen, x2, method='Newton-CG',
               jac=rosen_der, hess=rosen_hess,
               options={'xtol': 1e-8, 'disp': True})
```
